```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Grab the csv file and create a path for it
Neighborhood_path = './Neighborhood_Grouped.csv'

# read the original dataset (in comma separated values format) into a DataFrame
Neighborhood_Info = pd.read_csv(Neighborhood_path,index_col='Neighborhood')
```

```python
#checking contents 

Neighborhood_Info.head()
```

```python
# Plotting the everything calculated in the Neighborhood

Neighborhood_Info.plot(figsize=(20, 10))
```

```python
plt.figure(figsize=(20,10))
```

```python
# Just testing different kinds of graphs

Neighborhood_Info.plot.area(x="Neighborhood", y="mean", alpha=0.5)
```

```python
# Make a list of just the Mean and drop the other columns

N_mean = Neighborhood_Info.drop(['count','std','min','25%','50%','75%','max'], axis=1)
```

```python
# Build a bar Graph with Neighborhoods (y-axis) and the mean of the neighbhorhoods (x-axis)

Neighborhood_Info.plot.bar(x="Neighborhood", y="mean",figsize=(20, 10))
```

```python
# Build a bar Graph with Neighborhoods (y-axis) and the Count of the neighbhorhoods (x-axis)
# This Graph shows the number of properties per Neighborhood.

Neighborhood_Info.plot.bar(x="Neighborhood", y="count",figsize=(20, 10))
```

```python
# Created subplots and graphed for every calculated column in Neighborhood.

sub = Neighborhood_Info.plot.bar(x="Neighborhood",figsize=(30, 15), subplots=True)
```

```python
# Dropped percentage calculations.

N_stats = Neighborhood_Info.drop(['25%','50%','75%'], axis=1)
```

```python
# Created subplots and graphed for the columns, Count, Mean, Standard Deviation, Min, MAx in Neighborhood.

sub_plot = N_stats.plot.bar(x="Neighborhood",figsize=(20, 10), subplots=True)
```
